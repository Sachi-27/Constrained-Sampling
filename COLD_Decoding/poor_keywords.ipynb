```python
# from transformers import GPT2Tokenizer
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import os
import torch
import torch.nn.functional as F

# word = tokenizer.convert_ids_to_tokens(14299)
os.environ["CUDA_VISIBLE_DEVICES"] = "3"
device = "cuda"

model = GPT2LMHeadModel.from_pretrained(
    'gpt2-xl', output_hidden_states=True,
    resid_pdrop=0, embd_pdrop=0, attn_pdrop=0, summary_first_dropout=0)

model.to(device)

# Load tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-xl')

possible_sentences = [ #16 sentences
    "I am ",
    "Let us meet on the weekend for ",
    "He is going on a ",
    "The cat sat ",
    "Poor people are ",
    "The weather is so ",
    "I want to do ",
    "Please maintain ",
    "The world is very ",
    "She likes to make ",
    "Flowers bloom in ",
    "Sun is shining brightly this ",
    "Mother is cooking ",
    "My teacher doesn't like to ",
    "I was awarded zero marks for "
]

```

```python
required_indices_list = []

for sentence in possible_sentences:
    x_ = tokenizer(sentence, return_tensors='pt').to(device)
    outputs = model(**x_)
    preds = F.softmax(outputs.logits, dim=-1)
    # indices = [x.item() for x in list(torch.where(preds[0][0]<0.00001)[0])]
    ordered_indices = torch.argsort(preds[0][0])

    # Of these indices, add the lowest 100
    required_indices = []
    for i in ordered_indices:
        if len(required_indices) < 1000:
            required_indices.append(i.item())
        else:
            break

    required_indices_list.append(required_indices)
```

```python
index_freq = {}
for indices in required_indices_list:
    for index in indices:
        if index in index_freq:
            index_freq[index] += 1
        else:
            index_freq[index] = 1
```

```python
# Print the indices with their frequencies in descending order
sorted_index_freq = sorted(index_freq.items(), key=lambda x: x[1], reverse=True)
for index, freq in sorted_index_freq:
    print(f"{index}: {freq}")
```

```python
words = tokenizer.convert_ids_to_tokens([index for index, freq in sorted_index_freq])
print(words)
```
